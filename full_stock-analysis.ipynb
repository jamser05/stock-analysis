```python
import pandas_datareader.data as web
import matplotlib.pyplot as plt
import yfinance as yf
import datetime
import numpy as np
import pandas as pd
```

```python
telsa_ticker_symbol = 'TSLA'
ford_ticker_symbol = 'F'
gm_ticker_symbol = 'GM'

start = datetime.datetime(2012, 1, 1)
end = datetime.datetime(2024, 8, 5)
```

```python
tesla = yf.download(telsa_ticker_symbol, start, end)
ford = yf.download(ford_ticker_symbol, start, end)
gm = yf.download(gm_ticker_symbol, start, end)
```

```python
tesla.to_csv('Telsa_Stock.csv')
ford.to_csv('Ford_Stock.csv')
gm.to_csv('GM_Stock.csv')
```

```python
fig, ax = plt.subplots(figsize=(10, 10))

ax.plot(
    tesla['Open'],
    c='blue',
    label='Tesla'
)

ax.plot(
    gm['Open'],
    c='red',
    label='GM'
)

ax.plot(
    ford['Open'],
    c='pink',
    label='Ford'
)

plt.legend()
plt.title('Stock Prices')
plt.show()
```

```python
tesla['Total Traded'] = tesla['Open'] * tesla['Volume']
gm['Total Traded'] = gm['Open'] * gm['Volume']
ford['Total Traded'] = ford['Open'] * ford['Volume']
```

Candlestick Charts

```python
from mpl_finance import candlestick_ohlc
from matplotlib.dates import DateFormatter, date2num, WeekdayLocator, DayLocator, MONDAY

ford_reset = ford.loc['2012-01':'2012-01'].reset_index()

ford_reset['date_ax'] = ford_reset['Date'].apply(
    lambda date: date2num(date)
)

ford_values = [tuple(vals) for vals in ford_reset[['date_ax', 'Open', 'High', 'Low', 'Close']].values]

mondays = WeekdayLocator(MONDAY)
alldays = DayLocator()
weekFormmatter = DateFormatter('%b %d')
dayFormatter = DateFormatter('%d')

fig, ax = plt.subplots()

candlestick_ohlc(ax, ford_values, width=0.6, colorup='g', colordown='r')
```

Daily Percentage Change: r_t = (p_t/p_t-1) - 1

```python
tesla['returns'] = (tesla['Close']/tesla['Close'].shift(1)) - 1
ford['returns'] = (ford['Close']/ford['Close'].shift(1)) - 1
gm['returns'] = (gm['Close']/gm['Close'].shift(1)) - 1
```

Votality

```python
fig, ax = plt.subplots(figsize=(13, 6))

ax.hist(tesla['returns'], color='red', label='TESLA', bins=50, alpha=0.5)
ax.hist(ford['returns'], color='blue', label='FORD', bins=50, alpha=0.5)
ax.hist(gm['returns'], color='purple', label='GM', bins=50, alpha=0.5)
plt.legend()
```

```python
tesla['returns'].plot(kind='kde', label='TESLA')
gm['returns'].plot(kind='kde', label='GM')
ford['returns'].plot(kind='kde', label='Ford')

plt.legend()
plt.show()
```
